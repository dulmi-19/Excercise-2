```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("spider.png")
assert img is not None
fig, ax = plt.subplots(1,6)
fig.set_figwidth(20)
ax[0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
ax[0].set_title("Original Image")


list_gamma = np.array([0.2,0.8,1,1.2,2])
fig2,axis = plt.subplots(1,5)
fig2.set_figwidth(20)
for i , gamma in enumerate(list_gamma):
    transform = np.array([(p/255)**gamma*255 for p in range (0,256)]).astype(np.uint8)
    transformed = cv.LUT(img, transform)
    ax[i+1].imshow(cv.cvtColor(transformed, cv.COLOR_BGR2RGB))
    ax[i+1].set_title("$\gamma$ = "+ str(gamma))
    axis[i].plot(transform)
    axis[i].set_title("$\gamma$ = "+ str(gamma))
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im =  cv.imread("spider.png")
assert im is not None

t1=np.linspace(0,100,50)
t2=np.linspace(100,255,150)
t3=np.linspace(255,255,56)

transform = np.concatenate((t1,t2,t3), axis = 0).astype(np.uint8)
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_aspect('equal')
assert len(transform) == 256

im2= cv.LUT(im,transform)
fig2, axis = plt.subplots(1,2)
fig2.set_figwidth(20)
axis[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
axis[0].set_title("Original Image")
axis[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB))
axis[1].set_title("transformed image")
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im =  cv.imread("shells.tif", cv.IMREAD_GRAYSCALE)
assert im is not None

histo = cv.calcHist([im],[0],None,[256],[0,256])
im_equalized = cv.equalizeHist(im)
equalized_histo = cv.calcHist([im_equalized],[0],None,[256],[0,256])

fig, ax = plt.subplots(1,4,constrained_layout=True)
fig.set_figwidth(20)
ax[0].plot(histo)
ax[0].set_title("Original Image Histogram")
ax[1].plot(equalized_histo)
ax[1].set_title("equalized Image Histogram")

ax[2].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[2].set_title("Original Image")
ax[3].imshow(im_equalized, cmap='gray', vmin=0, vmax=255)
ax[3].set_title("equalized Image")

plt.show()
```
